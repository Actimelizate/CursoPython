## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """ SELECT Launchsite FROM SPACEXTBL GROUP BY 1   """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
query2 = """ SELECT Launchsite FROM SPACEXTBL WHERE Launchsite LIKE '%CCS%' """
df  = pd.read_sql(query2, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query3 = """ SELECT SUM(PayloadMass) AS MASA, Customers AS CLIENTE FROM SPACEXTBL WHERE Customers = 'NASA (CRS)' """
df  = pd.read_sql(query3, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query4 = """ SELECT AVG(PayloadMass) AS MEDIA, Serial AS SERIE FROM SPACEXTBL WHERE Serial LIKE '%B1%'  """
df  = pd.read_sql(query4, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query5 = """ SELECT MIN(Date) AS Fecha, LandingOutcome AS ATR_EXIST FROM SPACEXTBL WHERE LandingOutcome = 'True RTLS' """
df  = pd.read_sql(query5, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query6 = """ SELECT LandingOutcome AS ATR_EXIST, PayloadMass AS MASA FROM SPACEXTBL WHERE LandingOutcome = 'True ASDS' AND PayloadMass BETWEEN 4000 AND 6000 
ORDER BY MASA DESC"""
df  = pd.read_sql(query6, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query7 = """ SELECT LandingOutcome, COUNT(LandingOutcome) AS ATR_EXIST FROM SPACEXTBL GROUP BY LandingOutcome """
df  = pd.read_sql(query7, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query8 = """ SELECT MAX(PayloadMass) FROM SPACEXTBL """
df  = pd.read_sql(query8, con = engine )
df
```

```python
query8 = """ SELECT  Serial,PayloadMass FROM SPACEXTBL WHERE PayloadMass = (SELECT MAX(PayloadMass) FROM SPACEXTBL) """
df  = pd.read_sql(query8, con = engine )
df
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query9 = """ SELECT LandingOutcome AS ATERRIZAJE, Serial AS SERIE, Date AS FECHA FROM SPACEXTBL 
WHERE Date = 2015 AND LandingOutcome NOT LIKE '%True%'  """
df  = pd.read_sql(query9, con = engine )
df
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
query10 = """ SELECT COUNT(LandingOutcome) AS ATERRIZAJE, LandingOutcome FROM SPACEXTBL 
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY LandingOutcome ORDER BY ATERRIZAJE DESC"""
df  = pd.read_sql(query10, con = engine )
df
```
